```python
#operations
tensor=torch.tensor([1,2,3])
print(tensor +10)#tensor.add(tensor,10)
print(tensor -10)#tensor.sub(tensor,10)
print(tensor *10)#tensor.mul(tensor ,10)
print(tensor*tensor)#elementwise
print(tensor@tensor)#matrix multiplication
print(torch.matmul(tensor,tensor))
```

```python
tensor_A=torch.tensor([[1,2],[3,4],[5,6]],dtype=torch.float32)
tensor_B=torch.tensor([[7,10],[8,11],[9,12]],dtype=torch.float32)
#torch.matmul(tensor_A,tensor_B)
torch.matmul(tensor_A,tensor_B.T)#torch.mm() or @
print(tensor_A.shape)
print(tensor_B.shape)
print(tensor_A)
print(tensor_B)
```
